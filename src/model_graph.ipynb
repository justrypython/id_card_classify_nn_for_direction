```python
import os
import numpy as np
from keras.optimizers import SGD
from convnets import convnet
```

```python
sgd = SGD(lr=0.01, decay=5e-4, momentum=0.9, nesterov=True)
```

```python
model = convnet('alexnet', heatmap=False)
model.compile(optimizer=sgd, loss='mse', metrics=['accuracy'])
```

```python
from keras.utils.vis_utils import plot_model
from IPython.display import Image

plot_model(model, to_file="model_alexnet_227.png", show_shapes=True)
Image('model_alexnet_227.png')
```

```python
model = convnet('alexnet', heatmap=True)
model.compile(optimizer=sgd, loss='mse', metrics=['accuracy'])
```

```python
from keras.utils.vis_utils import plot_model
from IPython.display import Image

plot_model(model, to_file="model_alexnet_227_heap.png", show_shapes=True)
Image('model_alexnet_227_heap.png')
```
